```python
dili = cv.imread("./image/dili.jpg")
```

```python
plt.imshow(dili[:,:,::-1])
```

```python
b,g,r = cv.split(dili)
```

```python
plt.imshow(b,cmap=plt.cm.gray)
```

```python
img2 = cv.merge((b,g,r))
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
gray = cv.cvtColor(dili,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap=plt.cm.gray)
```

```python
hsv = cv.cvtColor(dili,cv.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```
