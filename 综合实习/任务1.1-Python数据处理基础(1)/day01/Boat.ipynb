```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取数据
data_src = 'data/titanic-data.csv'
df = pd.read_csv(data_src, header=0)
```

```python
# 查看数据集的基本信息
print(df.info())
```

```python
# 查看数据的摘要信息
print(df.describe())
```

```python
# 查看前几行数据
print(df.head())
```

```python
# 计算总体生存率
survived_rate = float(df['Survived'].sum()) / df['Survived'].count()
print('survived_rate: ', survived_rate)
```

```python
# 按舱位计算生存率
x = [df[(df.Pclass == 1)]['Pclass'].size, df[(df.Pclass == 2)]['Pclass'].size, df[(df.Pclass == 3)]['Pclass'].size]
y = [df[(df.Pclass == 1) & (df.Survived == 1)]['Pclass'].size, df[(df.Pclass == 2) & (df.Survived == 1)]['Pclass'].size, df[(df.Pclass == 3) & (df.Survived == 1)]['Pclass'].size]

print(f'1 Pclass number: {x[0]}, 1 Pclass survive: {y[0]}, 1 Pclass survive rate: {float(y[0]) / x[0]}')
print(f'2 Pclass number: {x[1]}, 2 Pclass survive: {y[1]}, 2 Pclass survive rate: {float(y[1]) / x[1]}')
print(f'3 Pclass number: {x[2]}, 3 Pclass survive: {y[2]}, 3 Pclass survive rate: {float(y[2]) / x[2]}')
```

```python
# 按舱位绘制生存率柱状图
Pclass_survived_rate = df.groupby('Pclass')['Survived'].mean()
Pclass_survived_rate.plot(kind='bar')
plt.title('Pclass Survived Rate')
plt.show()
```

```python
# 按性别计算生存人数
male_survived = df[(df.Sex == 'male')]['Survived'].sum()
female_survived = df[(df.Sex == 'female')]['Survived'].sum()

print('male survive:', male_survived)
print('female survive:', female_survived)
```

```python
# 按性别绘制生存率柱状图
Sex_survived_rate = df.groupby('Sex')['Survived'].mean()
Sex_survived_rate.plot(kind='bar')
plt.title('Sex Survived Rate')
plt.show()
```

```python
# 按年龄组计算生存率
age_clean_data = df[df.Age.notnull()]
ages = np.arange(0, 81, 5)
age_cut = pd.cut(age_clean_data.Age, ages)
age_cut_grouped = age_clean_data.groupby(age_cut)
age_Survival_Rate = age_cut_grouped['Survived'].mean()
age_Survival_Rate.plot(kind='bar')
plt.title('Age Group Survived Rate')
plt.show()
```

```python
# 按舱位和性别绘制生存率柱状图
Pclass_Sex_survived_rate = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
Pclass_Sex_survived_rate.plot(kind='bar')
plt.title('Pclass and Sex Survived Rate')
plt.show()
```
