```python
import pandas as pd
```

```python
name = ["file_num", "chars"]

data = pd.read_csv("./GenPics/labels.csv", names=name, index_col="file_num")
```

```python
labels = []
for label in data["chars"]:
    
    tmp = []
    
    for letter in label:
        
        tmp.append(ord(letter) - ord("A"))
        
    labels.append(tmp)
```
