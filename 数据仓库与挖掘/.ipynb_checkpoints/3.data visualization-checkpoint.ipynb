```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#创建图像区域，指定比例（可以不指定）
plt.figure(figsize=(8,6))
```

```python
#绘制线型图
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
plt.plot(x,y,'bp--')
plt.show()
```

```python
plt.figure()
data = pd.DataFrame([[1,2,3],[2,3,4]])
data.plot(kind='line')
plt.show()
```

```python
#饼图
plt.figure()
labels =["A","B","C","D"]
sizes = [15,30,45,10]
colors = ['yellow','green','lightskyblue','lightcoral']
explode =[0,0.1,0,0]
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
#条形图
plt.figure()
x = np.random.randn(1000)
plt.hist(x,10)
plt.show()
```

```python
#箱线图
plt.figure()
D = pd.DataFrame([x,x+1]).T
D.plot(kind='box')
plt.show()
```

```python
#误差条形图
error = np.random.randn(10)
y = pd.Series(np.sin(np.arange(10)))
y.plot(yerr=np.abs(error))
plt.figure()
plt.show()
```
